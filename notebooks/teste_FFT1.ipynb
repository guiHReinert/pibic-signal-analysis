```python
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import signal
```

```python
plt.rcParams['figure.figsize'] = [15,15]
plt.rcParams['font.size'] = 20

N = 1000
# sample spacing
T = 1/10000

x = T*np.arange(N) # Define x domain

y1 = 3*np.sin(2*np.pi*150*x) 
y2 = 2*np.cos(3*np.pi*250*x)

y = y1+y2

plt.subplot(3, 1, 1)
plt.title('f(x) = 3sin(300 pi x)')
plt.plot(x, y1)
plt.grid()

plt.subplot(3, 1, 2)
plt.title('g(x) = 2cos(750 pi x)')
plt.plot(x,y2)
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(x,y)
plt.grid()

plt.tight_layout()
plt.show()

# Transformada
plt.rcParams['figure.figsize'] = [15,15]

yhat = np.fft.fft(y)

plt.subplot(3,1,1)
plt.plot(yhat, label='fft')
plt.legend()
plt.grid()

xf = 1/(N*T)*np.arange(N)
xhat = np.fft.fftfreq(N,T)

#fig, ax = plt.subplots()
plt.subplot(3,1,2)
plt.plot(xf[0:100], np.abs(yhat)[0:100], label='xf')
plt.plot(xhat[0:100],  np.abs(yhat)[0:100], label='fftfreq') # np.abs() retorna o módulo de ()
plt.legend()
plt.grid()

plt.subplot(3,1,3)
plt.title('Transformada inversa')
iyhat = np.fft.ifft(yhat)
plt.plot(iyhat)
plt.grid()

plt.tight_layout()
```

```python
plt.rcParams['figure.figsize'] = [15,10]
plt.rcParams['font.size'] = '20'

# Função quadrada

n = 1000
T = 1/10000
x = np.linspace(0,10,n)
a1 = 10
y = a1 * signal.square(2 * np.pi * x)
            
plt.subplot(2,2,1)
plt.title('Função quadrada')
print('X shapes:', x.shape, '; y shapes:', y.shape)
plt.plot(y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

four = np.fft.fft(y)
freqs = np.fft.fftfreq(n, T)
ifour = np.fft.ifft(four)

plt.subplot(2,2,2)
plt.title('FFT inversa')
plt.plot(ifour)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,(3,4))
plt.title('FFT')
plt.plot(np.abs(freqs), np.abs(four))
plt.grid()
plt.xlabel('Frequência')
plt.ylabel('Amplitude')

#plt.ylim((6366.5,6367))

plt.tight_layout()
print(signal.find_peaks(np.abs(four), 6000))
```

```python
plt.rcParams['figure.figsize'] = [15,10]
plt.rcParams['font.size'] = '20'

# Função triangular

n = 1000
T = 1/10000
a2 = 10
y = a2 * signal.sawtooth(2 * np.pi * np.linspace(0,10,n), 0.5)

yhat = np.fft.fft(y)
iyhat = np.fft.ifft(yhat)
yfreqs = np.fft.fftfreq(n,T)

plt.subplot(2,2,1)
plt.title('Função triangular')
plt.plot(y, c='green')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,2)
plt.title('FFT inversa')
plt.plot(iyhat, c='green')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,(3,4))
plt.title('FFT ')
plt.plot(np.abs(yfreqs), np.abs(yhat), c='green')
plt.grid()
plt.xlabel('Frequência')
plt.ylabel('Amplitude')
plt.xlim((0,2000))

plt.tight_layout()
print(signal.find_peaks(np.abs(yhat), 4000))
```

```python
plt.rcParams['figure.figsize'] = [15,10]
plt.rcParams['font.size'] = '20'

# Função dente de serra

n = 1000
T = 1/10000
a3 = 10
y = a3 * signal.sawtooth(2 * np.pi * np.linspace(0,10,n))
        
yhat = np.fft.fft(y)
iyhat = np.fft.ifft(yhat)
yfreqs = np.fft.fftfreq(n,T)

plt.subplot(2,2,1)
plt.title('Função dente de serra')
plt.plot(y, c='red')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,2)
plt.title('FFT inversa')
plt.plot(iyhat, c='red')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,(3,4))
plt.title('FFT ')
plt.plot(np.abs(yfreqs), np.abs(yhat), c='red')
plt.grid()
plt.xlabel('Frequência')
plt.ylabel('Amplitude')

plt.tight_layout()
print(signal.find_peaks(np.abs(yhat), 3000))
```
