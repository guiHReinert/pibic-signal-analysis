```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Load dataset
H = np.loadtxt('curva_original.txt')
               
t = H[:,0]
y = H[:,1]

N = len(y)

yhat = np.fft.fft(y)
f = np.fft.fftfreq(N,t[1]-t[0])

plt.plot(f[0:50], np.abs(yhat[0:50]))
plt.grid()
plt.show()

plt.plot(f[0:50], (np.abs(yhat[0:50])**2))
plt.grid()
plt.show()
```

```python
V = np.fft.ifft(yhat)

plt.plot(t, V.real, 'b-', t, V.imag, 'r--')
plt.legend(('real', 'imaginary'))

plt.show()
```

```python
plt.plot(t, V.real, 'b-', t, y, 'r--')
plt.legend(('IFFT', 'original'))

plt.show()
```
