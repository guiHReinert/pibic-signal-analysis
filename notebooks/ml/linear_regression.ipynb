```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (15,10)
plt.rcParams['font.size'] = 13

# f(x) = y = ax + b
a = 3;                       
x = np.arange(-10, 10)
b = np.random.normal(0, 2, len(x))

# Simulação do ruído
y1 = a*np.array(x) + b

plt.figure(figsize=(15,7.5))
plt.plot(x, np.array(x)*a, label='Reta')
plt.plot(x, y1, 'x')
plt.legend()
plt.grid()

# REGRESSÃO...
x = x.reshape(-1,1) # Tranformando a matriz-linha em uma matriz-coluna
U, S, Vh = np.linalg.svd(x, full_matrices='False')

# Invertendo S
Si = np.array([1/S[0]])

# Arrumando a ordem de (Vh)(S^-1)
VhS = Vh @ Si
VhS = [VhS.tolist()]
for column in range((U.T).shape[0]-1):
    VhS[0].append(0)
    
VhS = np.array(VhS)
    
# Arrumando a ordem de y1
y1 = y1.reshape(-1,1)

# Produzindo a dita regressão por mínimos quadrados
xtilde = VhS @ np.conj(U).T # y = ax + b
xtilde = xtilde @ y1
xtilde = xtilde[0][0]

plt.plot(x,xtilde*x,'b--', label='Regressão')
plt.legend()
```
