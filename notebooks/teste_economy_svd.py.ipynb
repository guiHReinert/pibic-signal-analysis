```python
X = np.random.rand(10, 7)                                   # create random data matrix
U, S, V = np.linalg.svd(X, full_matrices=True)              # full SVD
Uhat, Shat, Vhat = np.linalg.svd(X, full_matrices=False)    # economy SVD
```
