```python
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import signal
```

```python
plt.rcParams['figure.figsize'] = [15,15]
plt.rcParams['font.size'] = '15'

# Função quadrada

n = 2000
T = 1/100
x = np.arange(n) * T
a1 = 10
y = a1 * signal.square(2 * np.pi * x)
            
plt.subplot(4,1,1)
plt.title('Função quadrada')
plt.plot(y)
plt.grid()

four = np.fft.fft(y)
freqs = np.fft.fftfreq(n, T)
ifour = np.fft.ifft(four)

plt.subplot(4,1,2)
plt.title('Transformada inversa')
plt.plot(ifour)
plt.grid()

plt.subplot(4,1,3)
plt.title('Transformadas')
plt.plot(x, np.abs(four), label='x')
plt.plot(np.abs(freqs), np.abs(four), label='fftfreq')
plt.legend()
plt.grid()

plt.subplot(4,1,4)
plt.title('plot do mano')
plt.plot(x, freqs, color='purple')
plt.grid()

plt.tight_layout()
```

```python
plt.rcParams['figure.figsize'] = [15,17]
plt.rcParams['font.size'] = '15'

# Função quadrada

n = 2000
T = 1/100
x = np.arange(n) * T
a = 10
y = a * signal.square(2 * np.pi * x)
            
plt.subplot(4,2,(1,2))
plt.title('Função quadrada')
plt.plot(x, y)
plt.grid()

four = np.fft.fft(y)
freqs = np.fft.fftfreq(n, T)
ifour = np.fft.ifft(four)

plt.subplot(4,2,(3,4))
plt.title('Transformada inversa')
plt.plot(x, ifour)
plt.grid()

plt.subplot(4,2,(5,5)) # Com negativos
plt.plot(x, four, label='x', color='b')
plt.grid()
plt.legend()

plt.subplot(4,2,(6,6)) # Módulo
plt.plot(x, np.abs(four), label='x', color='m')
plt.grid()
plt.legend()

plt.subplot(4,2,(7,7)) # Com negativos
plt.plot(np.abs(freqs), four, label='fftfreq', color='r')
plt.legend()
plt.grid()

plt.subplot(4,2,(8,8)) # Módulo
plt.plot(np.abs(freqs), np.abs(four), label='fftfreq', color='orange')
plt.grid()
plt.legend()


plt.tight_layout()
```
