```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os, PIL

plt.rcParams['figure.figsize'] = [16, 8] # Change it's styles and dimensions as well


A = imread(os.path.join('images/test.jpg')) # Merge the internal path str in only one
X = np.mean(A, -1); # Uma imagem RGB são três matrizes. O -1 significa que está fazendo uma média sobre as cores

img = plt.imshow(X)
img.set_cmap('gray') # Set grayscale
plt.axis('off') # remove the eixos(axis)
plt.show()
```

```python
U, S, VT = np.linalg.svd(X,full_matrices=False) # Receiving It's economy SVD, obviously

S = np.diag(S) # Receiving the S diagonal only
Xfull = U @ S @ VT

name = line = column = 0

# Construct approximate image

Xapprox1 = U[:,:1] @ S[0:1,:1] @ VT[:1,:]
Xapprox2 = U[:,:5] @ S[0:5,:5] @ VT[:5,:]
Xapprox3 = U[:,:20] @ S[0:20,:20] @ VT[:20,:]
Xapprox4 = U[:,:100] @ S[0:100,:100] @ VT[:100,:]

plt.figure(name, figsize=(10, 10)) # Entitula e dimesiona a próxima imagem
plt.subplots_adjust(wspace=0.01, hspace=0.2)

plt.subplot(2,2,1)
plt.imshow(Xapprox1)
plt.title('r = 1')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(Xapprox2)
plt.title('r = 5')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(Xapprox3)
plt.title('r = 20')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(Xapprox4)
plt.title('r = 100')
plt.axis('off')

#plt.tight_layout(w_pad=10)
plt.set_cmap('gray')
plt.show()
```

```python
# Economy SVD plot test
plt.figure('full')
img = plt.imshow(Xfull)
img.set_cmap('gray')
plt.axis('off')
plt.title('\"Full\" Economy SVD')
plt.show()

# Singular Values Graphics
plt.figure(1, figsize=(6,4))
plt.semilogy(np.diag(S))
plt.title('Valores singulares')
plt.xlabel('r')
plt.show()

plt.figure(2, figsize=(6,4))
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Somatório dos valores singulares')
plt.xlabel('r')
plt.show()
```

```python
# X times X*
plt.figure('XX*', figsize=(8.25,8.25))
XX = X @ np.conj(X).T
img = plt.imshow(XX)
img.set_cmap('gray')
print(XX.shape)
plt.title('XX*')
plt.axis('off')
plt.show()

# X* times X
plt.figure('X*X', figsize=(5.2,5.2))
XX = np.conj(X).T @ X
img = plt.imshow(XX)
img.set_cmap('gray')
print(XX.shape)
plt.title('X*X')
plt.axis('off')
plt.show()
```
