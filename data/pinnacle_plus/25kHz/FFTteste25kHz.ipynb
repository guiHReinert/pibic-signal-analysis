```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = [15,20]
plt.rcParams['font.size'] = 13

y_exp = np.loadtxt('curva_original.txt')

t = y_exp[:,0]
y = y_exp[:,1]

fft = np.fft.fft(y)
freqs = np.fft.fftfreq(len(y), t[1] - t[0])
ifft = np.fft.ifft(fft)

# Limitador: -1 para todos os termos
n = -1

plt.subplot(3,1,1)
plt.plot(t[:n], y[:n], t[:n], ifft[:n])
plt.legend(('Original', 'iFFT'))
plt.grid()

plt.subplot(3,1,2)
plt.plot(np.abs(freqs)[:n], np.abs(fft)[:n], label='FFT')
plt.legend()
plt.grid()

plt.subplot(3,1,3)
plt.plot(np.abs(freqs)[:n], label='Frequências')
plt.legend()
plt.grid()
```
